```python
# Main Consolidating all files
count=0
data=pd.DataFrame()
for file in os.listdir('/content/drive/MyDrive/08_Sales/01_Sales_Reports/C_Zomato/New Sheet Format files After Oct 29th 2022'):

  get_excel=pd.ExcelFile('/content/drive/MyDrive/08_Sales/01_Sales_Reports/C_Zomato/New Sheet Format files After Oct 29th 2022/'+file)
  count+=1
  read_excel=pd.read_excel(get_excel,'Order level',skiprows=range(6))
  read_excel.insert(0,'Source_files',file,allow_duplicates=False,)
  data=pd.concat([data,read_excel])

print(count)


```

### ADDING STANDARD COLUMS 

```python
#'B_Discount Given'
data['B_Discount Given']=data['Restaurant discount [Promo]']+ data['Restaurant discount [Salt, Freebie, Pro & other discounts]']
```

```python
#H_Gross Revenue Incl Taxes
data['H_Gross Revenue Incl Taxes'] = data['Subtotal (items total)'] - data['B_Discount Given'] + data['Total GST collected from customers']+data['Packaging charge']+data['Delivery charge']
```

```python
# I_Gross Revenue Excl Taxes
data['I_Gross Revenue Excl Taxes'] =data['Subtotal (items total)'] - data['B_Discount Given'] - data['Total GST collected from customers']+data['Packaging charge']+data['Delivery charge']
```

```python
# J_Commission

data['J_Commission']=data['Commission value \n[ (10) * (11) ]']+data['Taxes on Zomato fees\n[(12)+(13) +(14)] * 18%']
```

```python
# K_Long Distance Fees
data['K_Long Distance Fees']=0
```

```python
# L_Advertising Fees
data['L_Advertising Fees']=0
```

```python
# F_Rounded off
data['F_Rounded off']=0
```

```python
# F_Waived off
data['F_Waived off']=0
```

```python
# O_Other Deductions
data['O_Other Deductions']=data['Payment mechanism fee']+data['Logistics charge']
```

```python
#@title  creating synthetic columns
data["A_Subtotal"]=data["Subtotal (items total)"]

data["C_Packaging Charges"]=data["Packaging charge"]

data["D_Delivery Service Charges"]=data["Delivery charge"]

data["E_Tax Collected"]=data["Total GST collected from customers"]

data["E_GST Paid by Aggregator"]=data["GST retained by Zomato - Amount of tax paid by Zomato under section 9(5)"]

data["E_GST Paid by Merchant"]=data["GST to be paid by Restaurant partner to Govt."]

data["M_TDS on Revenue"]=data["TDS 194O amount"]

data["N_Customer Refunds"]=data["Net Additions \n(cancellation refund)"]

```

### %% Order Date formating from here
bold text

```python
# spliting order date column
order_datetime=list(data['Order Date'].str.split(' '))
```

```python
order_datetime[1]
```

```python
# appedning date from order date colm
order_date_seperated=[]
for date in order_datetime:
  order_date_seperated.append(date[0])
#print(order_date_seperated)
```

```python
# spliting order date to day month and year
splitted_order_date_seperated=[]
for date in order_date_seperated:
  splitted_order_date_seperated.append(date.split("-"))
splitted_order_date_seperated[1]
```

```python
# now creating new orddate colm
new_orderdate=[]
for new_date in splitted_order_date_seperated:
  new_orderdate.append(new_date[2]+"-"+new_date[1]+"-"+new_date[0])
new_orderdate[1]
```

```python
# Pushing the new order to dataframe i.e data 
data['Order_date'] = new_orderdate
```
